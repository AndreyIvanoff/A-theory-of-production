## Exploring Cobb and Douglas' Data with Standard Visualization Library

Let's inspect the data from Cobb and Douglas's (Cobb, Charles W., and Paul H. Douglas. "A theory of production." (1928), https://msuweb.montclair.edu/~lebelp/CobbDouglasProdAER1928.pdf) work using the standard visualization library.

```python
cb_data=[
    [1899,100,100,100],
    [1900,107,105,101],
    [1901,114,110,112],
    [1902,122,118,122],
    [1903,131,123,124],
    [1904,138,116,122],
    [1905,149,125,143],
    [1906,163,133,152],
    [1907,176,138,151],
    [1908,185,121,126],
    [1909,198,140,155],
    [1910,208,144,159],
    [1911,216,145,153],
    [1912,226,152,177],
    [1913,236,154,184],
    [1914,244,149,169],
    [1915,266,154,189],
    [1916,298,182,225],
    [1917,335,196,227],
    [1918,366,200,223],
    [1919,387,193,218],
    [1920,407,193,231],
    [1921,417,147,179],
    [1922,431,161,240]
]
cb_data = np.array(cb_data)

K = cb_data[:,1] # K: Relative Capital Stock, 1899=100
L = cb_data[:,2] # L: Relative Number of Workers, 1899=100
P = cb_data[:,3] # P: Index of Manufactures
t = cb_data[:,0] # years
```

```python
plot.plot(t,K,'--.k', label='K: Relative Capital Stock, 1899=100')
plot.plot(t,L,'--.r', label='L: Relative Number of Workers, 1899=100')
plot.plot(t,P,'--.b', label='P: Index of Manufactures')
plot.legend(loc='upper left')
```

```python
import numpy as np
xtickvals = np.append(np.arange(1899,1922,2),1922)
xticktexts = ['1899','1901','03','05','07','1909','11','13','15','17','1919','21','22']
```
